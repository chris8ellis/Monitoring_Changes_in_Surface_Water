# Intro to tensorflow.keras 

```python
from tensorflow.keras.preprocessing.image import array_to_img
```

```python
array_to_img(img, data_format='channels_first')
```

## Reverse array in vertical plane 

```python
from tensorflow.keras.backend import reverse
```

```python
rev = reverse(img, 1) # reverse in axes 1 (vertical)
```

```python
array_to_img(rev, data_format='channels_first')
```
